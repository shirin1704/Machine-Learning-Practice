```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
```

```python
data = pd.read_csv('/content/used_cars.csv')
data
```

```python
X = data.iloc[:, 1:]
Y = data.iloc[:, 0]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 1)
```

```python
bagging = BaggingRegressor(random_state = 1)
random_forest = RandomForestRegressor(random_state = 1)
gradient_boosting = GradientBoostingRegressor(random_state = 1)
adaboost = AdaBoostRegressor(random_state = 1)
```

```python
bagging.fit(X_train, Y_train)
random_forest.fit(X_train, Y_train)
gradient_boosting.fit(X_train, Y_train)
adaboost.fit(X_train, Y_train)
```

```python
bagging_pred = bagging.predict(X_test)
random_forest_pred = random_forest.predict(X_test)
gradient_boosting_pred = gradient_boosting.predict(X_test)
adaboost_pred = adaboost.predict(X_test)
```

```python
print("Bagging:", round(bagging.score(X_test, Y_test), 4))
print("Random Forest:", round(random_forest.score(X_test, Y_test), 4))
print("Gradient Boosting:", round(gradient_boosting.score(X_test, Y_test), 4))
print("AdaBoost:", round(adaboost.score(X_test, Y_test), 4))
```

```python
from sklearn.model_selection import GridSearchCV

X = data.iloc[:, 1:]
Y = data.iloc[:, 0]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 1)
```

```python
adaboost = AdaBoostRegressor(random_state = 1)

param_grid = {'n_estimators': [100, 500, 1000],
              'learning_rate': [0.5, 1, 2]}

grid_search = GridSearchCV(estimator=adaboost, param_grid=param_grid, cv=4)

grid_search.fit(X_train, Y_train)

Y_pred = grid_search.predict(X_test)
```

```python
print("Score:", grid_search.score(X_test, Y_test))
print("Best n: ", grid_search.best_params_['n_estimators'])
print("Best lr: ", grid_search.best_params_['learning_rate'])
```
