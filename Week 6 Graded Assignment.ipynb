```python
from sklearn.datasets import fetch_openml
from sklearn.linear_model import Perceptron
import numpy as np
```

```python
mnist = fetch_openml('mnist_784', version = 1, as_frame = False)
mnist
```

```python
X, Y = mnist.data, mnist.target.astype(int)
```

```python
X_train, Y_train = X[:10000], Y[:10000]
X_test, Y_test = X[10000:12000], Y[10000:12000]
```

```python
X_train_5 = X_train[Y_train == 5]
X_train_3 = X_train[Y_train == 3]
X_test_5 = X_test[Y_test == 5]
X_test_3 = X_test[Y_test == 3]
```

```python
X_train = np.vstack((X_train_5, X_train_3))
X_test = np.vstack((X_test_5, X_test_3))
```

```python
Y_train = np.hstack((np.ones(len(X_train_5)), -np.ones(len(X_train_3))))
Y_test = np.hstack((np.ones(len(X_test_5)), -np.ones(len(X_test_3))))
```

```python
perceptron = Perceptron(max_iter = 100, eta0=1, shuffle = True, fit_intercept = True, random_state = 42)
perceptron.fit(X_train, Y_train)
```

```python
Y_pred = perceptron.predict(X_test)
```

```python
(Y_train == 1).sum()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(Y_test, Y_pred)
tn = cm[0, 0]
fp = cm[0, 1]
fn = cm[1, 0]
tp = cm[1, 1]
```

```python
from sklearn.metrics import recall_score, precision_score, accuracy_score

a1 = accuracy_score(Y_test, Y_pred)
p1 = precision_score(Y_test, Y_pred)
r1 = recall_score(Y_test, Y_pred)
```

```python
accuracy = (tn + tp)/(tn + tp + fn + fp)
precision = tp/(tp + fp)
recall = tp/(tp + fn)
```

```python
perceptron_2 = Perceptron(max_iter = 100, eta0=1, shuffle = False, fit_intercept = True, random_state = 42)
perceptron_2.fit(X_train, Y_train)
Y_pred_2 = perceptron_2.predict(X_test)
```

```python
a2 = accuracy_score(Y_test, Y_pred_2)
p2 = precision_score(Y_test, Y_pred_2)
r2 = recall_score(Y_test, Y_pred_2)
```
