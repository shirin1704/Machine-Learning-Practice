```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
import pandas as pd
```

```python
data = load_wine()
X = data.data
Y = data.target

X = pd.DataFrame(data = X, columns=data.feature_names)
Y = pd.DataFrame(data = Y, columns=['Target'])

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=1)

classifier = DecisionTreeClassifier(random_state = 1)

param_grid = {'criterion' : ['entropy', 'gini'],
              'splitter' : ['random', 'best'],
              'min_samples_leaf' : [2,4,6,8,10],
              'max_depth' : [3,4,5,6],
              }

grid_search = GridSearchCV(estimator = classifier, param_grid = param_grid, cv = 4)
grid_search.fit(X_train, Y_train)

Y_pred = grid_search.predict(X_test)

grid_search.score(X_test, Y_test)

grid_search.best_params_
```

```python
from sklearn.datasets import load_diabetes
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = load_diabetes()

X = pd.DataFrame(data = data.data, columns=data.feature_names)
Y = pd.DataFrame(data = data.target, columns=['Target'])

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=1)

regressor = DecisionTreeRegressor(criterion='squared_error', splitter = 'random', max_leaf_nodes=10, random_state = 1)
regressor.fit(X_train, Y_train)

Y_pred = regressor.predict(X_test)

regressor.score(X_train, Y_train)

regressor.score(X_test, Y_test)

plt.figure(figsize=(25,20))
plot_tree(regressor, filled=True, feature_names=data.feature_names)
plt.show()
```

```python
from sklearn.datasets import load_diabetes
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
```

```python
data = load_diabetes()

X = pd.DataFrame(data = data.data, columns=data.feature_names)
Y = pd.DataFrame(data = data.target, columns=['Target'])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=1)
```

```python
regressor = DecisionTreeRegressor(max_depth = 4, random_state = 1)
```

```python
param_grid = {'criterion' : ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
              'splitter' : ['random', 'best'],
              'max_features' : [4,5,6,'sqrt', 'log2'],
              'ccp_alpha' : [0.001,0.01,0.02,0.05]
}

grid_search = GridSearchCV(estimator = regressor, param_grid = param_grid, cv = 4)
grid_search.fit(X_train, Y_train)

Y_pred = grid_search.predict(X_test)

grid_search.score(X_test, Y_test)
```
